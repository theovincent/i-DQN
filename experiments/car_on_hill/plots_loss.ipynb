```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import json

experiment_name = "q70_35_N50000_b100_b20_f4_t1_lr4_5_b_t"
list_bellman_iterations_scope = [1, 20]
seed = 5
p = json.load(open(f"figures/{experiment_name}/parameters.json"))
```

```python
for bellman_iterations_scope in list_bellman_iterations_scope:
    ifqi_loss = np.load(f"figures/{experiment_name}/iFQI/{bellman_iterations_scope}_L_{seed}.npy")
    plt.plot(ifqi_loss, label=f"k={bellman_iterations_scope}")
    plt.title(f"iFQI training loss, {bellman_iterations_scope * p['fitting_steps_per_bellman_iteration']} training steps per forward moves\n {p['target_update_per_gradient_step']} target update frequency")
    plt.legend()
    _ = plt.show()
```
