```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import json

experiment_name = "q100_h100_n250_M10000_b32_e100_b66_s1_pu_g300_t1_lr4_iu"
list_bellman_iterations_scope = [1, 11, 33, 66]
seed = 1
p = json.load(open(f"figures/{experiment_name}/parameters.json"))
```

```python
for bellman_iterations_scope in list_bellman_iterations_scope:
    ifqi_loss = np.load(f"figures/{experiment_name}/iDQN/{bellman_iterations_scope}_L_{seed}.npy")
    
    for idx_epoch in range(0, p["n_epochs"], 15):
        plt.plot(ifqi_loss[idx_epoch], label=f"epoch {idx_epoch + 1}")

    plt.title(f"iDQN training loss for {bellman_iterations_scope} ks, {bellman_iterations_scope * p['gradient_steps_per_bellman_iteration']} training steps per forward moves\n {p['target_updates_per_gradient_step']} target update frequency")
    plt.legend()
    _ = plt.show()
```
