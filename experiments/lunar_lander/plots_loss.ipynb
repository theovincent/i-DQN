```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import json

experiment_name = "q100_h100_n1000_M10000_b32_e10_b66_s1_pu_e1000_g300_t10_lr4_iu"
list_bellman_iterations_scope = [1, 3, 11]
seed = 1
p = json.load(open(f"figures/{experiment_name}/parameters.json"))
```

```python
for bellman_iterations_scope in list_bellman_iterations_scope:
    ifqi_loss = np.load(f"figures/{experiment_name}/iDQN/{bellman_iterations_scope}_L_{seed}.npy")
    plt.plot(ifqi_loss, label=f"k={bellman_iterations_scope}")
    plt.title(f"iDQN training loss, {bellman_iterations_scope * p['gradient_steps_per_bellman_iteration']} training steps per forward moves\n {p['target_updates_per_gradient_step']} target update frequency")
    plt.legend()
    _ = plt.show()
```
