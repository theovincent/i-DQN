```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import json

experiment_name = "test/Breakout"
list_bellman_iterations_scope = [10]
seed = 11
p = json.load(open(f"figures/{experiment_name.split('/')[0]}/parameters.json"))
```

```python
for bellman_iterations_scope in list_bellman_iterations_scope:
    ifqi_loss = np.load(f"figures/{experiment_name}/iDQN/{bellman_iterations_scope}_L_{seed}.npy")
    
    for idx_epoch in range(0, p["n_epochs"], 5):
        plt.plot(ifqi_loss[idx_epoch], label=f"epoch {idx_epoch + 1}")

    plt.title(f"iDQN training loss for {bellman_iterations_scope} ks")
    plt.legend()
    _ = plt.show()
```
